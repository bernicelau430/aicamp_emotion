```
import os
import sys
import cv2
from google.colab import drive
drive.mount('/content/gdrive')

emotion = 'poker' # important
  
img_folder = os.path.join(os.getcwd(), 'gdrive', 'My Drive', 'images')
emotion_folder = os.path.join(img_folder, emotion)
index = open(os.path.join(emotion_folder, 'index.txt'), "a")

for img_file in os.listdir(emotion_folder):
    if img_file.endswith((".png", ".jpg")):
        cascade_file = os.path.join(img_folder, 'haarcascade_frontalface_default.xml')
        face_cascade = cv2.CascadeClassifier(cascade_file)
        # eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
        img = cv2.imread(os.path.join(emotion_folder, img_file))
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)

        # TODO break this line apart
        label = open(os.path.join(emotion_folder, img_file.split(".")[0] + ".txt"), "w")
        for (x, y, w, h) in faces:
            img_h, img_w = img.shape[:2]
            center_x = (x + w / 2) / img_w
            center_y = (y + h / 2) / img_h
            rel_w = w / img_w
            rel_h = h / img_h
            label.write("0 {} {} {} {}".format(center_x, center_y, rel_w, rel_h))
        if len(faces) > 1:
          index.write("More than 1 face detected in {}\n".format(img_file))
          print("More than 1 face detected in {}".format(img_file))
        elif len(faces) == 0:
          index.write("No faces detected in {}\n".format(img_file))        
          print("No faces detected in {}".format(img_file))
        label.close()
        # TODO add if statement to take eyes into account when recognizing faces
        #     cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
        #     roi_gray = gray[y:y + h, x:x + w]
        #     roi_color = img[y:y + h, x:x + w]
        #     eyes = eye_cascade.detectMultiScale(roi_gray)
        #     for (ex, ey, ew, eh) in eyes:
        #         cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)
    else:
      print('uh oh {}'.format(img_file))
cv2.destroyAllWindows()
```
